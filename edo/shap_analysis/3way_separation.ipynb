```python
# colour assignment is universal
cols = {(0, ):'tab:blue', (1, ):'tab:orange', (0, 1):'tab:green', 0:'tab:blue', 1:'tab:orange'}    


def jaceks_method(f_vals, shap_vals):
    means = np.array([np.mean(shap_vals[f_vals==i]) for i in [0,1]])
    sigms = np.array([np.std(shap_vals[f_vals==i]) for i in [0,1]])
    sigms_v = sigms + 0.01  # non-zero variance for visualisation purposes
    
    left_bar = means-sigms_v
    length = 2*sigms_v
    
    return zip(left_bar, length), means, sigms_v


def gauss(x, miu, sigma):
    s_part = 1/(sigma*np.sqrt(2*np.pi))
    e_part = -0.5 * np.power((x-miu)/sigma, 2)
    return s_part * np.exp(e_part)


def meet_point(miu1, sigma1, miu2, sigma2):
    # gdzie przecinaja sie dwa gaussy
    sigma1sq = np.power(sigma1, 2)
    sigma2sq = np.power(sigma2, 2)
    a = 0.5 * ((1/sigma2sq) - (1/sigma1sq))
    b = (miu1/sigma1sq) - (miu2/sigma2sq)
    c = np.log(sigma2) - np.log(sigma1) + 0.5 * ((np.power(miu2,2)/sigma2sq) - (np.power(miu1,2)/sigma1sq))
    if np.isclose(a, 0):
        # równianie pierwszego rzędu
        x = -c/b
        return x, x
    delta = np.power(b, 2) - 4*a*c
    delta_sq = np.sqrt(delta)
    x1 = (-b-delta_sq)/(2*a)
    x2 = (-b+delta_sq)/(2*a)
    
    return x1, x2

def utnij(x, minimum, maximum):
    if x < minimum:
        print(f"{x} -> {minimum}")
        return minimum
    elif x > maximum:
        print(f"{x} -> {maximum}")
        return maximum
    return x

def draw2way(f_vals, shap_vals, threshold, majorities, purities, n_correct, title_suff=''):
    validate_shapes(f_vals, shap_vals, classes_order=None)
    title = "2-way separation"+title_suff

    # scatter plot preparation
    sorted_indices = np.argsort(shap_vals)
    x = shap_vals[sorted_indices]
    y_c = Counter(x)
    y = [i for k in sorted(y_c.keys()) for i in reversed(range(y_c[k]))]

    # regions preparation
    margin = 0.05*(max(shap_vals)-min(shap_vals))
    # l_region = (min(shap_vals)-margin, thresholds[0]-min(shap_vals)+2*margin)
    start = min(shap_vals)-margin
    end = max(shap_vals)+margin
    
    shaps_over_thr = shap_vals[shap_vals > threshold]
    # threshold equal to biggest shap value?
    if len(shaps_over_thr) == 0:
        thr = end
    else:   
        thr = (threshold+min(shaps_over_thr))/2 if not np.isinf(threshold) else start
    
    l_region = (start, thr-start)
    r_region = (thr, end-thr)
    
    region_colours = tuple(cols[m] for m in reversed(majorities))

    # actual drawing
    fig, ax = plt.subplots(figsize=(15,5))
    
    # jacek
    jacek_bars, means, sigms = jaceks_method(f_vals, shap_vals)
    mp1, mp2 = meet_point(means[0], sigms[0], means[1], sigms[1])
    mp1, mp2 = utnij(mp1, min(x), max(x)), utnij(mp2, min(x), max(x))
    plt.stem([mp1, mp2], [0.9*max(y), 0.9*max(y)], linefmt='g', basefmt='w', markerfmt='w',
             bottom=0.1, use_line_collection=True)
    xx = np.linspace(start, end, 1000)
    g1, g2 = [gauss(xx, mean, sigma) for mean, sigma in zip(means, sigms)]
    norm = max(y)/max(np.concatenate((g1, g2))) if max(y) > 0 else 1
    for yy in [g1, g2]:
        plt.plot(xx, norm*yy)
    # for i, (l_bar, length) in enumerate(jacek_bars):
    #     shift = (1+i)*0.15*max(y)
    #     ax.broken_barh([(l_bar, length)], (shift+0.15*max(y), shift+max(y)*0.25),
    #               facecolors=cols[i], alpha=0.1)
    
    # my separation
    ax.broken_barh(list(reversed([l_region, r_region])), (-0.1*max(y), max(y)*1.2),
                   facecolors='w', edgecolors=region_colours, lw=2)
    plt.scatter(x, y, c=[cols[f] for f in f_vals[sorted_indices]], alpha=0.1)

    ax.annotate(f"{np.round(purities[0], 2)} ({n_correct[0]})", (start+0.5*(thr-start), 0.9*max(y)))
    ax.annotate(f"{np.round(purities[1], 2)} ({n_correct[1]})", (thr+0.5*(end-thr),  0.9*max(y)))
    
    plt.title(title)
    plt.show()


def small_inpure_cluster_criterion(majorities, n_correct, min_n=20):
    # if true: that's a valid feature, if false:it's donwgraded to 2way
    # min_n - minimal number of samples for a cluster to remain
    for m, n in zip(majorities, n_correct):
        if m==(0,1) and n < min_n:
            return False
    return True 


def cluster_remover_criterion(n_correct, min_n=20):
    return len(n_correct) - reduce(lambda x,y: x+1 if y<min_n else x, n_correct, 0)

    
def draw3way(f_vals, shap_vals, thresholds, majorities, purities, n_correct, title_suff=''):
    validate_shapes(f_vals, shap_vals, classes_order=None)
    title = f"3-way -> {cluster_remover_criterion(n_correct)}-way feature;{title_suff}"

    # scatter plot preparation
    sorted_indices = np.argsort(shap_vals)
    x = shap_vals[sorted_indices]
    y_c = Counter(x)
    y = [i for k in sorted(y_c.keys()) for i in reversed(range(y_c[k]))]

    # regions preparation
    margin = 0.05*(max(shap_vals)-min(shap_vals))
    # l_region = (min(shap_vals)-margin, thresholds[0]-min(shap_vals)+2*margin)
    start = min(shap_vals)-margin
    l_x = (thresholds[0]+min(shap_vals[shap_vals > thresholds[0]]))/2
    r_x = (thresholds[1]+min(shap_vals[shap_vals > thresholds[1]]))/2
    end = max(shap_vals)+margin
    l_region = (start, l_x-start)
    m_region = (l_x, r_x-l_x)
    r_region = (r_x, end-r_x)
    region_colours = tuple(cols[m] for m in reversed(majorities))

    # actual drawing
    fig, ax = plt.subplots(figsize=(15,5))
    
    # jacek
    jacek_bars, means, sigms = jaceks_method(f_vals, shap_vals)
    mp1, mp2 = meet_point(means[0], sigms[0], means[1], sigms[1])
    mp1, mp2 = utnij(mp1, min(x), max(x)), utnij(mp2, min(x), max(x))
    plt.stem([mp1, mp2], [0.9*max(y), 0.9*max(y)], linefmt='g', basefmt='w', markerfmt='w',
             bottom=0.1, use_line_collection=True)
    xx = np.linspace(start, end, 1000)
    g1, g2 = [gauss(xx, mean, sigma) for mean, sigma in zip(means, sigms)]
    norm = max(y)/max(np.concatenate((g1, g2))) if max(y) > 0 else 1
    for yy in [g1, g2]:
        plt.plot(xx, norm*yy)
    # for i, (l_bar, length) in enumerate(jacek_bars):
    #     shift = (1+i)*0.15*max(y)
    #     ax.broken_barh([(l_bar, length)], (shift+0.15*max(y), shift+max(y)*0.25),
    #               facecolors=cols[i], alpha=0.1)
    
    # my separation
    ax.broken_barh(list(reversed([l_region, m_region, r_region])), (-0.1*max(y), max(y)*1.2),
                   facecolors='w', edgecolors=region_colours, lw=2)
    
    plt.scatter(x, y, c=[cols[f] for f in f_vals[sorted_indices]], alpha=0.1)

    ax.annotate(f"{np.round(purities[0], 2)} ({n_correct[0]})", (start+0.5*(l_x-start), 0.9*max(y)))
    ax.annotate(f"{np.round(purities[1], 2)} ({n_correct[1]})", (l_x+0.5*(r_x-l_x),  0.8*max(y)))
    ax.annotate(f"{np.round(purities[2], 2)} ({n_correct[2]})", (r_x+0.5*(end-r_x),  0.9*max(y)))
    
    plt.title(title)
    plt.show()
    
    
def draw_each_solution(feature, n_way, cls_id):
    # select draw_func
    if 2 == n_way:
        draw_func = draw2way
    elif 3 == n_way:
        draw_func = draw3way
    else:
        raise NotImplementedError()
    
    # restrict to selected class
    if cls_id is not None:
        solutions = feature.well_separated(n_way)[cls_id]
        s_vals = feature._shap_values[cls_id,:]
    else:
        solutions = feature.well_separated(n_way)
        s_vals = feature._shap_values
    
    # make title suffix
    for sol in solutions:
        n_correct = np.sum(sol.n_correct)
        ratio = n_correct/feature._nsamples
        title_suff = f"    n_correct: {n_correct} ({int(100*ratio)}%)    score: {np.round(sol.score, 2)}"

        draw_func(feature._feature_values, s_vals, 
                  sol.thresholds, sol.majorities, sol.purities, sol.n_correct,
                  title_suff)    
      
    

def analyse_and_draw(ftr, cls_id=None):
    # ftr - Feature
    # cls_id in case of Task.CLASSIFICATION defines for which class the results are drawn
    # (analyses are done for all classes)
    print(ftr)
    # calculating separations
    ftr.well_separated(n_groups=2)
    ftr.well_separated(n_groups=3)
    
    # ftr.info()
    for way in [2, 3]:
        draw_each_solution(ftr, way, cls_id)        
    print('\n')
 
```
