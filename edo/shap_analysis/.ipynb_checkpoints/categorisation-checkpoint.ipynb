```python
# colour assignment is universal
cols = {(0, ):'tab:blue', (1, ):'tab:orange', (0, 1):'tab:green', 0:'tab:blue', 1:'tab:orange'}    


def min_max_cropper(x, minimum, maximum):
    if np.isinf(x):
        return minimum if x < 0 else maximum
    return x


def draw(f_vals, s_vals, regions, title_pref='', title_suff=''):
    validate_shapes(f_vals, s_vals, classes_order=None)
    title = title_pref+title_suff

    # scatter plot preparation
    sorted_indices = np.argsort(s_vals)
    x = s_vals[sorted_indices]
    y_c = Counter(x)
    y = [i for k in sorted(y_c.keys()) for i in reversed(range(y_c[k]))]
    dot_colours = [cols[f] for f in f_vals[sorted_indices]]
    
    # broken_barh plot preparation
    margin = 0.05*(max(s_vals)-min(s_vals))
    starts = tuple(min_max_cropper(r.start, min(s_vals)-margin, max(s_vals)+margin) for r in regions)
    ends =   tuple(min_max_cropper(r.end,   min(s_vals)-margin, max(s_vals)+margin) for r in regions)
    rectangles = tuple((s, max(0, e-s)) for s, e in zip(starts, ends))
    rect_colours = tuple(cols[r.majority] for r in regions)

    # actual drawing
    fig, ax = plt.subplots(figsize=(15,5))
    ax.broken_barh(list(reversed(rectangles)), (-0.1*max(y), max(y)*1.2),
                   facecolors='w', edgecolors=list(reversed(rect_colours)), lw=2)
    plt.scatter(x, y, c=dot_colours, alpha=0.1)
    
    for i, (reg, (start, width)) in enumerate(zip(regions, rectangles)):
        ax.annotate(f"{np.round(reg.purity, 2)} ({reg.n_correct})", (start+0.5*width,  (0.9-i*0.1)*max(y)))
    
    plt.title(title)
    plt.show()


def cluster_remover_criterion(n_correct, min_n=20):
    return len(n_correct) - reduce(lambda x,y: x+1 if y<min_n else x, n_correct, 0)


    
def restrict_to_selected_class(feature, func_name, kwargs, cls_id=None):
    solutions = getattr(feature, func_name)(**kwargs)
    s_vals = feature._shap_values
    if cls_id is not None:
        solutions = solutions[cls_id]
        s_vals = s_vals[cls_id,:]
    return solutions, s_vals


def draw_empty(feature, cls_id):
    _, s_vals = restrict_to_selected_class(feature, '__str__', {}, cls_id=cls_id)
    draw(feature._feature_values, s_vals, [], str(feature), '')


def draw_each_solution_ws(feature, n_way, cls_id):
    if 2 == n_way:
        title_pref_foo = lambda n_correct: "2-way separation"
    elif 3 == n_way:
        title_pref_foo = lambda n_correct: f"3-way -> {cluster_remover_criterion(n_correct)}-way feature"
    else:
        raise NotImplementedError()
    
    solutions, s_vals = restrict_to_selected_class(feature, 'well_separated', {'n_groups':n_way},
                                                   cls_id=cls_id)
    for sol in solutions:
        # make title suffix
        n_correct = np.sum(r.n_correct for r in sol.regions)
        ratio = n_correct/feature._nsamples
        title_pref = title_pref_foo([r.n_correct for r in sol.regions])
        title_suff = f"    n_correct: {n_correct} ({int(100*ratio)}%)    score: {np.round(sol.score, 2)}"

        draw(feature._feature_values, s_vals, sol.regions,
             title_pref, title_suff)    


def draw_hg(feature, threshold, metric, cls_id):
    title_pref = 'high gain'
    sol, s_vals = restrict_to_selected_class(feature, 'high_gain',
                                             {'gamma':threshold, 'criterion':metric},
                                             cls_id=cls_id)
    if np.isnan(sol.score) or sol.score <= 0:
        print("No high gain samples.")
        return
    title_suff = f"   score: {np.round(sol.score, 2)} {sol.params}"
    draw(feature._feature_values, s_vals, [sol.loss_region, sol.gain_region],
         title_pref, title_suff)
    

def draw_si(feature, threshold, metric, cls_id):
    title_pref = 'selectively important'
    sol, s_vals = restrict_to_selected_class(feature, 'selectively_important',
                                             {'miu':threshold, 'criterion':metric},
                                             cls_id=cls_id)
    if np.isnan(sol.score) or sol.score <= 0:
        print("Feature is not selectively important.")
        return
    title_suff = f"   score: {np.round(sol.score, 2)} {sol.params}"
    draw(feature._feature_values, s_vals, [sol.loss_region, sol.gain_region],
         title_pref, title_suff) 
        

def unimportance_statement(feature, threshold, metric, cls_id):
    sol, s_vals = restrict_to_selected_class(feature, 'unimportant',
                                             {'miu':threshold, 'criterion':metric},
                                             cls_id=cls_id)
    print(f"Feature unimportance score is {sol.score} (metric: {sol.params['metric']}).")


def analyse_and_draw(ftr, cls_id=None):
    # ftr - Feature
    # cls_id in case of Task.CLASSIFICATION defines for which class the results are drawn
    # (analyses are done for all classes)
    # print(ftr)
    # ftr.info()
    draw_empty(ftr, cls_id)
    
    # calculating separations
    ftr.well_separated(n_groups=2)
    ftr.well_separated(n_groups=3)
    ftr.high_gain(gamma=gamma, criterion='ratio')
    ftr.high_gain(gamma=gamma, criterion='absolute')
    ftr.selectively_important(miu=miu, criterion='ratio')
    ftr.selectively_important(miu=miu, criterion='absolute')
    ftr.unimportant(miu, criterion='ratio')
    ftr.unimportant(miu, criterion='absolute')
    
    
    for way in [2, 3]:
        draw_each_solution_ws(ftr, way, cls_id)
        
    for draw_func, threshold in zip((draw_hg, draw_si, unimportance_statement), (gamma, miu, miu)):
        for criterion in ['absolute', 'ratio']:
            draw_func(ftr, threshold, criterion, cls_id)
            
    print(f"")
            
    print('\n')
 
```

```python
def small_inpure_cluster_criterion(majorities, n_correct, min_n=20):
    # if true: that's a valid feature, if false:it's donwgraded to 2way
    # min_n - minimal number of samples for a cluster to remain
    for m, n in zip(majorities, n_correct):
        if m==(0,1) and n < min_n:
            return False
    return True 

def cluster_remover_criterion(n_correct, min_n=20):
    return len(n_correct) - reduce(lambda x,y: x+1 if y<min_n else x, n_correct, 0)
        

def draw(feature_vals, shap_vals, thresholds, majorities, purities, n_correct):
    # title = "OK feature"
    # title = title if small_inpure_cluster_criterion(majorities, n_correct) else "small inpure cluster -> 2way"
    title = f"{cluster_remover_criterion(n_correct)}-way feature"
    
    
    cols = {(0, ):'tab:blue', (1, ):'tab:orange', (0, 1):'tab:green', 0:'tab:blue', 1:'tab:orange'}
    
    # samples
    sorted_indices = np.argsort(shap_vals)
    
    x = shap_vals[sorted_indices]
    y_c = Counter(x)
    y = [i for k in sorted(y_c.keys()) for i in reversed(range(y_c[k]))]
    
    # regions
    margin = 0.05*(max(shap_vals)-min(shap_vals))
    # l_region = (min(shap_vals)-margin, thresholds[0]-min(shap_vals)+2*margin)
    start = min(shap_vals)-margin
    l_x = (thresholds[0]+min(shap_vals[shap_vals > thresholds[0]]))/2
    r_x = (thresholds[1]+min(shap_vals[shap_vals > thresholds[1]]))/2
    end = max(shap_vals)+margin
    l_region = (start, l_x-start)
    m_region = (l_x, r_x-l_x)
    r_region = (r_x, end-r_x)
    region_colours = tuple(cols[m] for m in reversed(majorities))
    
    fig, ax = plt.subplots(figsize=(15,5))
    ax.broken_barh(list(reversed([l_region, m_region, r_region])), (-0.1*max(y), max(y)*1.2),
                   facecolors='w',
                   edgecolors=region_colours,
                   lw=2)
    plt.scatter(x, y, c=[cols[f] for f in feature_vals[sorted_indices]], alpha=0.1)
    
    ax.annotate(f"{np.round(purities[0], 2)} ({n_correct[0]})", (start+0.5*(l_x-start), 0.9*max(y)))
    ax.annotate(f"{np.round(purities[1], 2)} ({n_correct[1]})", (l_x+0.5*(r_x-l_x),  0.8*max(y)))
    ax.annotate(f"{np.round(purities[2], 2)} ({n_correct[2]})", (r_x+0.5*(end-r_x),  0.9*max(y)))
    plt.title(title)
    plt.show()
    

def drawing(f_vals, s_vals, min_p=0.5):
    score, thr, maj, purs, ns = three_way_separation(f_vals, s_vals, min_p)
    print(f"score: {score}")
    for t, m, p, n in zip(thr, maj, purs, ns):
        # print(f"thr: {t}\tmaj: {m}")
        draw(f_vals, s_vals, t, m, p, n)
        
    return score, thr, maj
```
